### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure is the whole figure and can be seen as a canvas.
# Axes is the part of the canvas that the plot is attached on.
# Axis sets the scale and the limits and generate ticks and ticklabels.
# Artist is basically everything visible on the figure (including figure, axes and axis)
# En "figure" är själva ritduken, "axes" är området där du ritar dina diagram, "axis" är skalorna som markerar värdena
# och "artist" är de grafiska elementen som utgör ditt diagram.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
# The expected input is numpy arrays or objects that can be transformed to a numpy array (numpy.asarray).
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
#Creating a figure and axes
fig, ax = plt.subplots()

#Generate evenly spaced data from -4 to 4
x = np.linspace(-4, 4, 100)
y = x**2

#Giving the plot a title and labels to the axis.

ax.set_title('My first plot')
ax.set_xlabel('x-label')
ax.set_ylabel('y-label')

#Doing the plot by calling methods directly from the axes. (Object-oriented)
ax.plot(x, y)
```

```python
#Pyplot approach
plt.plot(x, y)
plt.title('My first plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
#Creating subplots
fig, axs = plt.subplots(1, 2, figsize = (12, 6), layout='constrained')
fig.suptitle('My subplots', size=30)
                        
axs[0].scatter(random_data_x, random_data_y)
axs[0].set_title('Scatter plot')

colors = ['purple', 'green', 'orange', 'yellow', 'lightgreen']

axs[1].bar(names, values, color=colors)
axs[1].set_title('Bar plot')

plt.show()
```
